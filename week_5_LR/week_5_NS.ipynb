
Later on in the assignment I can see that the independent variable is the TV advertising spending and the dependent variable is the product sales; this is because we're wanting to find if thre is a relationship between that of TV advertising spending and product sales. I'll use the sci-kit linear regression function.

In order to use the function on the two variables, I have to figure how to pass them as arguments. Looking into the docs, it's noted that "LinearRegression takes in its fit method arguments X, y, sample_weight"; this seems to imply that I would need to create objects x & y to represent my variables. The objects to be created are pandas dataframes, which provide a data structure similer to the dataset and are able for manuplation. In their creation, I have to use double brackets for the independent variable (the single feature) in order to create a 2-dimensional data frame, whereas for the dependent variable (the target) I only had to use single brackets for a 1-dimensional dataframe: why is that? Sci-kit always expects a 2-dim array for the independent value, where the two dimensions are n_samples (the rows) and n_features (the column, single here); interesting enough, the use of the method '.values' comes from numpy and converts the then created pandas dataframe into a numpy array.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/home/nate/NextCloud/Roam/Classes/Intro_to_ML/assignmnets/week_5/Advertising.csv')

X = df[['TV']].values
y = df['Sales'].values

model = LinearRegression()
model.fit(X, y)
y_pred=model.predict(X)

# Defining the figure, graph type & axes, and then ploting the regression line on top.
plt.figure (figsize=(10,6))
plt.scatter(X, y, color='blue', label='Ground Truth')
plt.plot(X, y_pred, color='red', linewidth=3, label='Regressin Lines of Best Fit')

#Adding metadata for better understanding
plt.title('Liner Regression: TV ads vs. Sales')
plt.xlabel('Spent on TV Ads')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)

plt.savefig('LRTS.png')
```


Looking over the plotted linear regression ontop of the scatterplot compairing what was spent on TV Ads and sales, I can see there is a kind of linear relationship between the amount spent on ads and the amount of sales, but towards the end the points start to become much more disperised. This seems to indicate that there might be, at some point, dimishing returns on the investing in TV Ads in order to boost sales.
