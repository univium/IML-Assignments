```
import pandas as pd
db=pd.read_csv("/home/nate/NextCloud/Roam/Classes/Intro_to_ML/assignmnets/data/JaipurFinalCleanData.csv")

agg_db=db.agg(
    {
        "min_pressure_2": ["mean", "median", "std", "var"],
        "rainfall": ["mean", "median", "std", "var" ],
    }
)

print(agg_db)
```
